## Red Wine Quality Converted to 3

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = 'wine-quality/winequality-red.csv'
features = ['fixed acidity', 'volatile acidity', 'citric acid',
           'residual sugar', 'chlorides', 'free sulfur dioxide',
           'total sulfur dioxide', 'density', 'pH', 'sulphates',
           'alcohol', 'quality']
data = pd.read_csv(path, header=None, names=features, sep=';')
data.head()
```

```python
data[data['quality'] > 6]
```

```python
data['quality'].describe()
```

```python
#3,4
#5,6
#7,8
data.loc[data['quality'] <=4, 'quality'] = 0 
data.loc[(data['quality'] >4) & (data['quality'] <=6), 'quality'] = 1
data.loc[(data['quality'] >6), 'quality'] = 2
```

```python
data['quality'].describe()
```

```python
x = data.iloc[:, 0:-1].values
y = data.iloc[:, -1].values.reshape(-1, 1)
from sklearn.preprocessing import normalize

x = normalize(x, norm='l2')
y = normalize(y, norm='l2')
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(x, y)
```

```python
from sklearn.svm import SVC

wine_svc = SVC(kernel='rbf', gamma='auto')
```

```python
wine_svc.fit(X_train, Y_train)
```

```python
test_predict = wine_svc.predict(X_test)

from sklearn.metrics import accuracy_score

accuracy_score(Y_test, test_predict)
```

## SVC kernel = rbf

```python
from sklearn.model_selection import cross_val_score

cross_val_score(wine_svc, x, y, cv=5)
```
